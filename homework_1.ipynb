### Exercise 1

```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
```

```python
def Verhulst(x, t, r, K):
  """
  Continuous-time logistic model
  """
  dxdt = r * x * (1 - x/K)
  return dxdt
```

```python
t = np.linspace(0, 10, num=101)
K_1 = 100
r = 1.5
x_0_1 = np.linspace(0, 200, num=101)
```

```python
x_1 = odeint(Verhulst, x_0_1, t, args=(r, K_1))
```

```python
plt.plot(t, x_1)
plt.xlabel('time')
plt.ylabel('y(t)')
plt.figure(figsize=(1,1), dpi=256)
plt.show()
```

### Exercise 2

```python
def Verhulst_Pearl(x, t, alpha, K):
  """
  Continuous-time logistic model
  """
  # np.log calculates the natural logarithm -- I checked!
  dxdt = alpha * x * np.log(K/x)
  return dxdt
```

```python
# I changed variables' subscript so they do not override Exercise's 1 values.
t = np.linspace(0, 10, num=101)
K_2 = 100
alpha = 1.5
x_0_2 = np.linspace(1, 200, num=100)
```

```python
x_0_2
```

```python
x_2 = odeint(Verhulst_Pearl, x_0_2, t, args=(alpha, K_2))
```

```python
plt.plot(t, x_2)
plt.xlabel('time')
plt.ylabel('y(t)')
plt.figure(figsize=(1,1), dpi=256)
plt.show()
```
